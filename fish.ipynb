# Menetukan model

### Menetukan Fitur dan Target

```python
print(df.columns)
```

```python
x = df[['Length1', 'Length2', 'Length3', 'Height',
       'Width']]
y = df['Species']
```

```python
# Merubah nama kolom
x = np.array(x)
y =np.array(y)
```

```python
x.shape
```

```python
y.shape
```

```python
# Split Data Training and testing
np.random.seed(42)
train_size = int(len(x) * 0.8)
idx = np.random.permutation(len(x))

# Data Latih dan Data Uji
x_train, x_test = x[idx[:train_size]], x[idx[train_size:]]
y_train, y_test = y[idx[:train_size]], y[idx[train_size:]]
```

```python
# Output nya
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Transpose (Ubah garis menjadi kolom)
x_transpose = x.T
```

```python
## Perkalian x.T dengan x
x_transpose_x = np.dot(x_transpose, x)
```

```python
x_transpose_x
```

```python
## Perkalian x.T dengan y
x_transpose_y = np.dot(x_transpose, y)
```

```python
x_transpose_y
```

```python
### Menghitung SLOPE and Intercept
weights = np.linalg.inv(x_transpose_x).dot(x_transpose_y)

intercept = (np.sum(y) - np.dot(weights, np.sum(x, axis=0))) / len(x)
```

```python
weights
```

```python
intercept
```

```python
# ### Menghitung Slopw dan intercept
### ValueError: operands could not be broadcast together with shapes (159,5) (159,
# slope = (len(x) * np.sum(x * y) - np.sum(y)) / (len(x) * np.sum(x**2) - np.sum()**2)

# intercept = (np.sum(y) - slope * np.sum(x)) / len(x)
```

```python
# Memprediksi
### round hanya bisa di gunakan untuk nilai tunggal, bukan array
###
y_pred = np.round(np.dot(x_test, weights) + intercept)
```

```python
y_pred
```

### Evaluasi model

```python
r2 = 1 - (np.sum((y_test - y_pred))**2 / np.sum((y_test - y_test.mean())**2))
r2
```

```python
# MSE Mean Squared Error
mse = np.mean((y_test - y_pred))
mse
```

```python
# MAE Mean Absulote Error
mae = np.mean(np.abs(y_test - y_pred)**2)
mae
```
